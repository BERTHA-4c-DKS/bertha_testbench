```python
import pandas as pd

db = pd.read_csv("results.csv", sep=",")

for c in db.columns:
    print("\"%s\""%(c))
#print(db)
```

```python
import numpy as np 

systems = list(set(db["System"].values))
scheds = list(set(db["Scheduling"].values))

#print(scheds)
#print(systems)

for sysin in systems:
    db1 = db[db["System"] == sysin]
    
    sys = sysin.replace(" ", "")

    print(sys)

    fp = open(sys+"_time.csv", "w")
    print("Scheduling , #t , Walltime , Speedup , Memory ")
    print("Scheduling , #t , Walltime , Speedup , Memory ", file=fp)
    
    for schedin in scheds:
        selected = db1[db1["Scheduling"] == schedin]

        sched = schedin.replace(" ", "")

        walltimepert = {}
        memorypert = {}
        for index, row in selected.iterrows():
            nomuft = row["NumOfThreads"]
            walltime = row["WallTime"]
            memory = row["Memory"]
         
            walltimepert[nomuft] = walltime
            memorypert[nomuft] = memory

        serial = 0.0
        for nomuft in walltimepert:
            avg = walltimepert[nomuft]
            speedup = 1.0
            if nomuft == 1:
                 serial = avg
            else:
                speedup = serial/avg
             
            print(" %7s , %3d , %10.3f , %5.2f , %15.3f "%(sched, nomuft, avg, speedup, \
                                                     memorypert[nomuft]))
            print(" %7s , %3d , %10.3f , %5.2f , %15.3f "%(sched, nomuft, avg, speedup, \
                                                     memorypert[nomuft]), file=fp)            
    fp.close()
    
```
